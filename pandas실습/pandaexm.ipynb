```python
df = pd.read_csv('netflix.csv')
df
```

```python
def re_duration(x):
    x = str(x)
    x = x.replace('h','시간')
    x = x.replace('m','분')
    return x

df['N_Duration'] = df['Duration'].apply(re_duration)
df
```

1.Age Rating 컬럼의 +를 제거하고 All 및 TV-MA 값은 0으로 만들어 숫자형태로 N_Age_Rating 컬럼에 할당하여라
2.N_Age_Rating 으로 그룹바이를 하여 나이별 개수를 파악하여라
3.Category 로 그룹바이를 하여 Category별 평균 나이를 구하여라
