```python
import pandas as pd

s_df = pd.read_csv('./static/stores.csv')
s_df = s_df.set_index('s_id', drop=False)
p_df = pd.read_csv('./static/products.csv')
p_df = p_df.set_index('p_id', drop=False)
iv_df = pd.read_csv('./static/inventory.csv')

j_df1=iv_df.set_index('s_id').join(s_df.set_index('s_id'))
j_df1
```

```python
j_df2=j_df1.set_index('p_id').join(p_df.set_index('p_id'))
j_df2
```

```python
jj_df=iv_df.set_index('p_id').join(p_df.set_index('p_id'))
jj_df2=jj_df.set_index('s_id').join(s_df.set_index('s_id'))
jj_df2
```

```python
s_list=[]
for s_id in jj_df2.index:
    s_list.append({'s_id': s_id, 'name': jj_df2.loc[s_id, 'name'],
                   'locate': jj_df2.loc[s_id, 'locate'],
                   'inventory': jj_df2.loc[s_id, 'product']})
print(s_list)
```
